```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn import preprocessing
import numpy as np
```

```python
trainingDF = pd.read_csv('/content/drive/MyDrive/B.Tech. Final Year Project/Offensive Language Detection/Datasets/m_dataset_21_9/dataset/train.csv')
testingDF = pd.read_csv('/content/drive/MyDrive/B.Tech. Final Year Project/Offensive Language Detection/Datasets/m_dataset_21_9/dataset/test.csv')
x_train = trainingDF['text'].to_list()
y_train = trainingDF['hate'].values
x_test = testingDF['text'].to_list()
y_test = testingDF['hate'].values
```

```python
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(x_train)
X_test = vectorizer.transform(x_test)
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
X_test_tfidf = tfidf_vectorizer.transform(x_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
classifier = LogisticRegression(max_iter=40224)
classifier.fit(X_train, y_train)
```

```python
classifier2 = LogisticRegression(max_iter=40224)
classifier2.fit(X_train_tfidf, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred_tfidf = classifier2.predict(X_test_tfidf)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy2 = accuracy_score(y_test, y_pred_tfidf)
print("Accuracy:", accuracy)
print("TF IDF Accuracy:", accuracy2)
```

```python
report = classification_report(y_test, y_pred)
report2 = classification_report(y_test, y_pred_tfidf)
print("Classification Report:\n", report)
print("Classification Report for TF IDF:\n", report2)
```
