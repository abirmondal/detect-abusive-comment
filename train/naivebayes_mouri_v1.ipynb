```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Load your data
train_data = pd.read_csv('/content/drive/MyDrive/B.Tech. Final Year Project/Offensive Language Detection/Datasets/m_dataset_21_9/dataset/train.csv')
test_data = pd.read_csv('/content/drive/MyDrive/B.Tech. Final Year Project/Offensive Language Detection/Datasets/m_dataset_21_9/dataset/test.csv')
val_data = pd.read_csv('/content/drive/MyDrive/B.Tech. Final Year Project/Offensive Language Detection/Datasets/m_dataset_21_9/dataset/val.csv')

# Preprocess data (assuming your dataset has 'text' and 'label' columns)
X_train = train_data['text']
y_train = train_data['hate']
X_test = test_data['text']
y_test = test_data['hate']
X_val = val_data['text']
y_val = val_data['hate']

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
X_val_tfidf = tfidf_vectorizer.transform(X_val)  # Transform validation data as well

# Naive Bayes with TF-IDF embeddings (Multinomial Naive Bayes)
nb_classifier = MultinomialNB()  # Use Multinomial Naive Bayes for TF-IDF
nb_classifier.fit(X_train_tfidf, y_train)
y_pred = nb_classifier.predict(X_test_tfidf)

# Calculate accuracy on test set
accuracy = accuracy_score(y_test, y_pred)

# Make predictions on validation set
y_val_pred = nb_classifier.predict(X_val_tfidf)

# Calculate accuracy on validation set
accuracy_val = accuracy_score(y_val, y_val_pred)

# print("Naive Bayes with TF-IDF (Test Set):")
# print(f"Accuracy: {accuracy}")  # Print accuracy on test set
# print(classification_report(y_test, y_pred))

print("Naive Bayes with TF-IDF (Validation Set):")
print(f"Accuracy: {accuracy_val}")  # Print accuracy on validation set
print(classification_report(y_val, y_val_pred))
```
